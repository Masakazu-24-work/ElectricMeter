```python
import pandas as pd
import networkx as nx
import openpyxl
import matplotlib.pyplot as plt

# 配電盤ツリー図のデータを読み込む
trees = pd.read_excel('配電盤ツリー図.xlsx', sheet_name='階層テーブル')
```

```python
inspected_result = pd.read_excel('ブレーカ調査/241203_ブレーカ調査.xlsx',header=1)
```

```python
parents = []
childrens = []
child_childrens = []

for index, row in inspected_result.iterrows():
    print(row['分電盤名'])
    if row['分電盤名'] in inspected_result['親分電盤\n(あれば)'].values:
        parent = row['分電盤名']
        children = inspected_result[inspected_result['親分電盤\n(あれば)'] == parent]['分電盤名'].values
        children = list(set(children))
        if len(children) != 0:
            for child in children:
                child_children = inspected_result[inspected_result['親分電盤\n(あれば)']==child]['分電盤名'].values
                if len(child_children) != 0:
                    child_children = list(set(children))
                    for child_child in child_children:
                        parents.append(parent)
                        childrens.append(child)
                        child_childrens.append(child_child)
                else:
                    parents.append(parent)
                    childrens.append(child)
                    child_childrens.append(None)
        else:
            parents.append(parent)
            childrens.append(None)
            child_childrens.append(None)
```

```python
# Create a dataframe from the lists parents, childrens, and child_childrens
df_combined = pd.DataFrame({
    'Parent': parents,
    'Children': childrens,
    'Child_Children': child_childrens
})

print(df_combined)
```
